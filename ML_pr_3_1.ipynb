```python
import numpy as np
import pandas as pd
import os

from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dense
```

```python
seed = 7
np.random.seed(seed)
```

```python
dataset = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/data/diabetes.csv',delimiter=',',skiprows=1)
dataset.shape
```

```python
X = dataset[:,:dataset.shape[1]-1]
Y = dataset[:,dataset.shape[1]-1]
```

```python
X[:2]
```

```python
Y[:2]
```

```python
model= keras.Sequential()
model.add(layers.Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
model.add(layers.Dense(8,kernel_initializer='uniform',activation='relu'))
model.add(layers.Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(  optimizer = 'adam', loss = 'BinaryCrossentropy', metrics = ['accuracy'])
```

```python
model.fit(
    X,Y,
    batch_size=10,
    epochs=150,
)
```

```python
scores = model.evaluate(X,Y)
print('%s: %.2f%%' % (model.metrics_names[1],scores[1]*100))
```

```python
X2 = dataset[:,0:8]
Y2 = dataset[:,8]
```

```python
kfold = StratifiedKFold(n_splits = 10, shuffle=True, random_state=seed)
cvscores = []
```

```python
for train, test in kfold.split(X2, Y2):

  model3= keras.Sequential()
  model3.add(layers.Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
  model3.add(layers.Dense(8,kernel_initializer='uniform',activation='relu'))
  model3.add(layers.Dense(1,kernel_initializer='uniform',activation='sigmoid'))

  model3.compile( loss = 'BinaryCrossentropy',  optimizer = 'adam', metrics = ['accuracy'])

  model3.fit(
    X2[train],Y2[train],
    batch_size=10,
    epochs=150,
    verbose=0
    )
  
  scores = model3.evaluate(X2[test],Y2[test],verbose=0)
  print('%s: %.2f%%' % (model.metrics_names[1],scores[1]*100))
  cvscores.append(scores[1]*100)
```

```python
print('%.2f%% (+/- %.2f%%)' % (np.mean(cvscores),np.std(cvscores)))
```
